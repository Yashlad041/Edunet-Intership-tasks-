```python
import pandas as pd

# Import CSV file
df = pd.read_csv("/content/popular_shows_dataset.csv")

# Display original data
print("Original Dataset:")
print(df)

# 1. head()
print("\n1. First 5 Rows:")
print(df.head())

# 2. tail()
print("\n2. Last 5 Rows:")
print(df.tail())

# 3. shape
print("\n3. Shape of Dataset:")
print(df.shape)

# 4. columns
print("\n4. Column Names:")
print(df.columns)

# 5. dtypes
print("\n5. Data Types:")
print(df.dtypes)

# 6. info()
print("\n6. Dataset Information:")
df.info()

# 7. describe()
print("\n7. Statistical Summary:")
print(df.describe())

# 8. isnull().sum()
print("\n8. Missing Values:")
print(df.isnull().sum())

# 9. sort_values()
print("\n9. Sorted by Revenue:")
print(df.sort_values(by="revenue", ascending=False))

# 10. sample()
print("\n10. Random 5 Rows:")
print(df.sample(5))
```
