```python

import pandas as pd
 Load the datasets

weather = pd.read_csv("/content/weather_features.csv")
energy = pd.read_csv("/content/energy_dataset.csv")


 Display first 5 rows

print("Weather Dataset")
print(weather.head())

print("\nEnergy Dataset")
print(energy.head())


 General Information

print("\nWeather Dataset Info")
print(weather.info())

print("\nEnergy Dataset Info")
print(energy.info())


General Analysis (Comments)


  Weather Dataset Analysis
 - Contains weather information of different cities.
 - Includes temperature, humidity, pressure, wind speed, clouds, etc.
 - Date/time column represents weather observation time.
- Each row represents weather conditions at a specific time.

  Energy Dataset Analysis
 - Contains electricity generation and market prices.
 - Includes energy produced from gas, coal, hydro, solar and wind.
- Price columns depend on electricity demand and production.
- Each row represents energy data for a particular date/time.


 Missing Values

print("\nMissing Values in Weather Dataset")
print(weather.isnull().sum())

print("\nMissing Values in Energy Dataset")
print(energy.isnull().sum())


 Handle Null Values



weather.fillna(weather.mean(numeric_only=True), inplace=True)
energy.fillna(energy.mean(numeric_only=True), inplace=True)


for col in weather.select_dtypes(include='object'):
    weather[col].fillna(weather[col].mode()[0], inplace=True)

for col in energy.select_dtypes(include='object'):
    energy[col].fillna(energy[col].mode()[0], inplace=True)

print("\nNull Values After Cleaning")
print(weather.isnull().sum())
print(energy.isnull().sum())


 Statistical Summary

print("\nWeather Summary")
print(weather.describe())

print("\nEnergy Summary")
print(energy.describe())


 Relationship Between Features


print("\nWeather Dataset Correlation")
print(weather.corr(numeric_only=True))

print("\nEnergy Dataset Correlation")
print(energy.corr(numeric_only=True))

weather.to_csv("weather_features_cleaned.csv", index=False)
energy.to_csv("energy_dataset_cleaned.csv", index=False)

print("\nCleaned datasets saved successfully!")

 SUMMARY


 1. Imported the Weather and Energy datasets using the Pandas library.
 2. Displayed the first few rows to understand the structure of the data.
 3. Checked dataset information such as column names, data types, and number of records.
 4. Identified missing (null) values in both datasets.
5. Handled missing values by:
   - Replacing null values in numeric columns with the column mean.
   - Replacing null values in categorical columns with the most frequent (mode) value.
 6. Generated statistical summaries using describe() to understand the data distribution.
 7. Analyzed relationships between numerical features using the correlation matrix.
 8. Added observations about how weather features and energy generation features are related.
 9. Saved the cleaned datasets as:
    - weather_features_cleaned.csv
    - energy_dataset_cleaned.csv

 Conclusion:
 The datasets were successfully explored, cleaned, and analyzed.
 Missing values were handled appropriately, relationships between important
 features were identified, and the cleaned datasets were saved for further
analysis or machine learning tasks
```
